```python
# Mini Project
## CSTR Design for a Second-Order Reaction
#A continuous stirred tank reactor (CSTR) is to be designed for a second-order irreversible reaction.
#The objective is to investigate how reactor volume varies with conversion and evaluate whether using two reactors in series provides better performance than a single reactor.
## Governing Equations
# -ra=kCa^2 
# Ca=Ca0*(1-X)
# V=Fa0*X/-ra
##Given data
# k=0.45 L/mol*min, Fa0=100 mol/min, Ca0=2mol/L
## Part A: Volume versus Conversion
k=0.45 #L/mol*min
Fa0=100 #mol/min
Ca0=2 #mol/L
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import pandas as pd
X=np.linspace(0,0.99,100) #Conversion range
ra=k*Ca0**2*(1-X)**2
V=Fa0*X/ra
#Plotting graph
plt.plot(X,V)
plt.xlabel('Conversion')
plt.ylabel('Reactor Volume (L)')
plt.title('Conversion v/s Reactor Volume curve')
plt.grid(True, color='grey', alpha=0.4)
plt.axvline(x=0.85,color='red',linestyle='--', linewidth='1.2',label='X=0.85')
plt.axvline(x=0.97,color='red',linestyle='--', linewidth='1.2',label='X=0.97')
plt.legend()
plt.show()
print('Reactor volume increases rapidly above approximately 90% conversion because the reaction rate decreases sharply as reactant concentration falls.')
## Part B: Calculation of Conversion for a given reactor volume
def Conversion(X):
    Target_Volume=950 #in L
    V=Target_Volume
    ra=k*Ca0**2*(1-X)**2
    return Fa0*X/ra-V
Ans=fsolve(Conversion, 0.85)
print(f'The Conversion rate for reactor volume of 950L is approximately {np.round(Ans[0],3)}')
#Date Table showing non-linear increase of Volume with increasing Conversion
Conversion_Viable=np.array(['85%','90%','95%','97%'])
Volumes=[]
for X in np.array([0.85,0.90,0.95,0.97]):
    ra=k*Ca0**2*(1-X)**2
    Volume=Fa0*X/ra
    Volumes.append(round(Volume,2))
Data=pd.DataFrame({'Conversion(%)':Conversion_Viable,'Reactor Volume(L)':Volumes})
print(Data)
## Part C: Analysis of two CSTR in series 
#Volume of both reactors is given 
V1=400 #in L
V2=400 #in L
def CSTR_Series(conversion):
    X1,X2=conversion
    ra1=k*Ca0**2*(1-X1)**2
    ra2=k*Ca0**2*(1-X2)**2
    return [Fa0*X1/ra1-V1,Fa0*(X2-X1)/ra2-V2]
Ans=fsolve(CSTR_Series,[0.75,0.8])
print(f'The conversion for only the first reactor is {np.round(Ans[0],2)} and the overall conversion achieved for both reactors is {np.round(Ans[1],2)}') 
##Part D: Economic Discussion and Conclusions
#Increasing conversion from 95% to 97% requires more than double the reactor volume. Unless downstream separation costs dominate, the marginal gain in conversion may not justify the additional capital investment.
#Two CSTRs in series achieve higher overall conversion than a single reactor of equal total volume because the reaction rate is higher in the first reactor where reactant concentration is greatest.
#Reactor volume increases nonlinearly with conversion.
#Above about 95% conversion, reactor size increases dramatically.
#Two CSTRs in series provide higher overall conversion than a single reactor of equal volume.
```
